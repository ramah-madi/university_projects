```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import numpy as np
import time
```

# Fearture Extraction

```python
coun_vect = CountVectorizer()
count_matrix = coun_vect.fit_transform(data['description'])
count_array = count_matrix.toarray()
df = pd.DataFrame(data=count_array,columns = coun_vect.get_feature_names())

X = df
y = data['category']
```

```python
from sklearn.tree import DecisionTreeClassifier


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
start_time = time.time()
DCTClassifier = DecisionTreeClassifier()
DCTClassifier.fit(X_train, y_train)
y_pred = DCTClassifier.predict(X_test)
D_acc = accuracy_score(y_test, y_pred)
end_time = time.time()
D_elapsed_time = end_time - start_time
print("accuracy = ",D_acc)
print("time of excution = ",D_elapsed_time)
```

```python
from sklearn.svm import SVC

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

start_time = time.time()
SVC = SVC()
SVC.fit(X_train, y_train)
y_pred = SVC.predict(X_test)
S_acc = accuracy_score(y_test, y_pred)

end_time = time.time()
S_elapsed_time = end_time - start_time
print("accuracy = ",S_acc)
print("time of excution = ",S_elapsed_time)
```

```python
from sklearn.neural_network import MLPClassifier


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

start_time = time.time()
MLPClassifier = MLPClassifier()
MLPClassifier.fit(X_train, y_train)
y_pred = MLPClassifier.predict(X_test)
M_acc = accuracy_score(y_test, y_pred)

end_time = time.time()
M_elapsed_time = end_time - start_time
print("accuracy = ",M_acc)
print("time of excution = ",M_elapsed_time) 
```
